# Sales Performance Analysis

## Objective
The goal of this project is to analyze sales data and identify trends, top-performing products, and regional performance.

```python
import pandas as pd

df = pd.read_csv("Sample_superstore.csv", encoding='latin1')
df.head()
```

## Data Understanding

```python
df.info()
```

```python
df.describe()
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

## Data Analysis

### Top 10 Products

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

### Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

## Data Visualization

```python
import matplotlib.pyplot as plt
```

### Sales by Region (Bar Chart)

```python
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

### Top 10 Products (Bar Chart)

```python
top_products.plot(kind='bar')
plt.title("Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

## Key Insights
- Sales vary across different regions
- West region shows higher performance
- Few products contribute more to total revenue
